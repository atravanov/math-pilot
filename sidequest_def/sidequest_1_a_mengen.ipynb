## Theorie

### Überblick der wichtigsten Mathematischen Mengen
Wir werden folgende Mengen in Zukunft verwenden:


- $\mathbb{N}$ — Menge der **natürlichen Zahlen**  
  $\mathbb{N} = \{0, 1, 2, 3, \dots\}$  
  oder ohne Null: $\mathbb{N}^+ = \{1, 2, 3, \dots\}$

- $\mathbb{Z}$ — Menge der **ganzen Zahlen**  
  $\mathbb{Z} = \{\dots, -2, -1, 0, 1, 2, \dots\}$

- $\mathbb{Q}$ — Menge der **rationalen Zahlen**  
  $\mathbb{Q} = \left\{ \frac{a}{b} \,\middle|\, a, b \in \mathbb{Z},\ b \ne 0 \right\}$

- $\mathbb{R}$ — Menge der **reellen Zahlen**  
  $\mathbb{R}$ enthält alle rationalen und irrationalen Zahlen

- $\mathbb{P}$ — Menge der **Primzahlen**  
  $\mathbb{P} = \{2, 3, 5, 7, 11, 13, 17, \dots\}$

- $\emptyset$ oder $\varnothing$ — die **leere Menge**  
  $\emptyset = \{\}$


### Teilmengen

Eine Menge $A$ ist eine **Teilmenge** von $B$ ($A \subseteq B$), wenn **alle Elemente von $A$ auch in $B$ enthalten sind**.

Beispiel (Mathematik):

$A = \{1, 2\}, \quad B = \{1, 2, 3, 4\} \quad \Rightarrow A \subseteq B$

Beispiel (Python):

```python
A = {1, 2}
B = {1, 2, 3, 4}
print(A.issubset(B))  # Ausgabe: True
```

### Vereinigungsmenge ($\cup$)

Die **Vereinigung** zweier Mengen $A$ und $B$ enthält **alle Elemente, die in $A$, $B$ oder beiden Mengen** vorkommen.

Mathematisch:  

$A \cup B = \{x \mid x \in A \lor x \in B\}$

Python:

```python
A = {1, 2, 3}
B = {3, 4, 5}
print(A | B)  # Ausgabe: {1, 2, 3, 4, 5}
```

### Schnittmenge ($\cap$)

Die **Schnittmenge** enthält **nur die Elemente, die in beiden Mengen vorkommen**.

Mathematisch:  
$A \cap B = \{x \mid x \in A \land x \in B\}$

Python:

```python
A = {1, 2, 3}
B = {2, 3, 4}
print(A & B)  # Ausgabe: {2, 3}
```

### Mengendifferenz ($\setminus$)

Die **Differenzmenge** $A \setminus B$ enthält **die Elemente, die in $A$, aber nicht in $B$ vorkommen**.

Mathematisch:

$A \setminus B = \{x \in A \mid x \notin B\}$

Python:

```python
A = {1, 2, 3}
B = {2, 3}
print(A - B)  # Ausgabe: {1}
```

### Symmetrische Differenz ($A \triangle B$)

Die **symmetrische Differenz** enthält **alle Elemente, die in genau einer der beiden Mengen vorkommen**, aber nicht in beiden.

Python:

```python
A = {1, 2, 3}
B = {3, 4, 5}
print(A ^ B)  # Ausgabe: {1, 2, 4, 5}
```

### Potenzmenge

Die **Potenzmenge** $\mathcal{P}(A)$ ist die Menge **aller Teilmengen** von $A$.

Beispiel:

Wenn $A = \{a, b\}$, dann ist

$$
\mathcal{P}(A) = \{\emptyset, \{a\}, \{b\}, \{a, b\}\}
$$

Python (mit `itertools`):

```python
import itertools

A = {'a', 'b'}
potenzmenge = [set(x) for i in range(len(A)+1) for x in itertools.combinations(A, i)]
print(potenzmenge)
```

### Zusammenfassung

| Operation              | Mathematisch      | Python-Syntax      |
|------------------------|-------------------|---------------------|
| Teilmenge              | $A \subseteq B$   | `A.issubset(B)`     |
| Vereinigung            | $A \cup B$        | `A \| B`             |
| Schnittmenge           | $A \cap B$        | `A & B`             |
| Differenz              | $A \setminus B$   | `A - B`             |
| Symmetrische Differenz | $A \triangle B$   | `A ^ B`             |
| Potenzmenge            | $\mathcal{P}(A)$     | `itertools.combinations` |

### Beispiel: Approximative Berechnung von $\pi$

```python
import matplotlib.pyplot as plt
import random

def pi_plot(num_points:int =10000) -> None:
    '''
    Führt eine Simulation zur Näherung von pi durch
    und zeigt eine Visualisierung des Resultats.
    '''
    inside_x, inside_y = [], []
    outside_x, outside_y = [], []

    inside_circle = 0

    for _ in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1
            inside_x.append(x)
            inside_y.append(y)
        else:
            outside_x.append(x)
            outside_y.append(y)

    approx_pi = 4 * inside_circle / num_points

    plt.figure(figsize=(6, 6))
    plt.scatter(inside_x, inside_y, color='green', s=1, label='im Kreis')
    plt.scatter(outside_x, outside_y, color='red', s=1, label='ausserhalb')
    plt.gca().set_aspect('equal')
    plt.title(f'Näherung von π ≈ {approx_pi:.5f} mit {num_points} Punkten')
    plt.legend()
    plt.grid(True)
    plt.show()

# Starte die Visualisierung
pi_plot(10000)
```

Der Algorithmus beruht auf den folgenden Tatsachen:

* Der Einheitskreis hat den Radius 1 und liegt im Quadrant $[0, 1] \times [0, 1]$
* Ein Punkt $(x, y)$ liegt im Kreis, wenn $x^2 + y^2 \leq 1$
* Das Verhältnis der Treffer im Kreis zu allen Punkten nähert $\frac{\pi}{4}$
* Daraus folgt:  
  $$
  \pi \approx 4 \cdot \frac{\text{Anzahl Punkte im Kreis}}{\text{Gesamtzahl Punkte}}
  $$

Das folgende Programm implementiert die Approximation von $\pi$ ohne Visualisierung:

```python
import random  # noqa: F811

def pi_approximation(num_points: int) -> float:
    """
    Approximiere den Wert von pi mit Hilfe der Monte-Carlo-Methode:
    Es werden zufällig Punkte in einem Einheitsquadrat generiert.
    Der Anteil der Punkte, die innerhalb des Einheitskreises liegen,
    wird zur Abschätzung von pi verwendet.

    Parameter:
    -----------
    num_points : int
        Anzahl der zufällig generierten Punkte

    Rückgabewert:
    -------------
    float
        Näherungswert von pi
    """
    inside_circle = 0

    for _ in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1

    return 4 * inside_circle / num_points

# Beispielaufruf:
approx_pi = pi_approximation(100_000)
print(f"Näherung von π mit 100.000 Punkten: {approx_pi}")
```
